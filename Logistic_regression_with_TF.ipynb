```
# Importing required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
import tensorflow as tf
from sklearn.model_selection import train_test_split
```

Getting the data.

```
from sklearn.datasets import load_iris
```

```
iris = load_iris()
```

```
iris_X, iris_y = iris.data[:-1,:], iris.target[:-1]
iris_y= pd.get_dummies(iris_y).values
```

```
train_x, test_x, train_y, test_y = train_test_split(iris_X, iris_y, test_size = 0.3, random_state = 12)
```

Our Logistic Regression simple equation :   Yhat = sigmodi( W * X + b)

```
#a = tf.matmul( X, weight , name = 'a')
#b = tf.add(a,bias, name = 'b')
#Yhat = tf.nn.sigmoid(b, name = 'Yhat')
```

```
num_features = train_x.shape[1]
num_labels = train_y.shape[1]
```

```
print(num_features)
print(num_labels)
```

If we go on a simple way of making model i.e. just by building the equation and then training it.

```
w = tf.Variable(tf.zeros([num_features,num_labels], dtype ='float32') )
b = tf.Variable(tf.zeros([1, num_labels], dtype='float32'))
```

```
print(w.dtype)
```

```
print(b.dtype)
```

```
print(w.shape)
print(b.shape)
```

```
a = tf.matmul(train_x, w, name='a')
```

```
a
```

```
a.shape
```

```
print(a)
```

```
b = tf.add(a, b ,name = 'b')
```

```
print(b.shape)
```

```
yhat = tf.nn.sigmoid(b, name = 'yhat')
```

```
yhat
```

```
yhat.shape
```

```
# if we want to some of the calculated values
init_op = tf.global_variables_initializer()
s = tf.Session()
s.run(init_op)
s.run(yhat[0:10])
```

```
s.run(a[0:10])
```

```
s.run(b[0:10])
```

```
# We need to close the session
s.close()
```

```
# now define the loss function 
loss = tf.reduce_mean(tf.square(yhat - train_y))
```

```
optimizer = tf.train.GradientDescentOptimizer(0.01)
```

```
train = optimizer.minimize(loss)
```

```
# Now to train the model we need to define a session and initialize all the variables and invoke a for loop

init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)
```

```
loss = []
t = []
for step in range(100):
    a, b, c , d = sess.run([train, loss , w, b])
    loss.append(b)
    if step % 5 ==0: 
        print(a, b , c ,d)
        t.append([c,d])
```
